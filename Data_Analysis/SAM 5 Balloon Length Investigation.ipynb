# SAM 5 Balloon Body Length Investigation
***

The first step in understanding this analysis is to understand what we're analysing. This machine produces 6 balloons every cycle some key points to consider are as follows, also reference Figure 1 for clarification; 

There are 3 arms; arm 1, arm 2 and arm 3
Each arm houses two moulds; left and right
For this study there were 2 thermocouples per mould; top and bottom

When defining variable the approach taken was;
Arm 1, Left, Top = A1LT, Arm 1, Left, Bottom = A1LB, Arm 2, Left, Top = A2LT....etc.


Figure 1

## Importing libraries and defining variables.
***

```python
import pandas as pd
import matplotlib.pyplot as plt
file_name = str(input('Include File Name for Analysis (include extension) '))
df = pd.read_excel(file_name)

a1lt = []
for i in df["A1LT"]:
    if i >= 30:
        a1lt.append(i)
a1lb = []
for i in df["A1LB"]:
    if i >= 30:
        a1lb.append(i)
a2lt = []
for i in df["A2LT"]:
    if i >= 30:
        a2lt.append(i)
a2lb = []
for i in df["A2LB"]:
    if i >= 30:
        a2lb.append(i)
a3lt = []
for i in df["A3LT"]:
    if i >= 30:
        a3lt.append(i)
a3lb = []
for i in df["A3LB"]:
    if i >= 30:
        a3lb.append(i)
a1rt = []
for i in df["A1RT"]:
    if i >= 30:
        a1rt.append(i)
a1rb = []
for i in df["A1RB"]:
    if i >= 30:
        a1rb.append(i)
a2rt = []
for i in df["A2RT"]:
    if i >= 30:
        a2rt.append(i)
a2rb = []
for i in df["A2RB"]:
    if i >= 30:
        a2rb.append(i)
a3rt = []
for i in df["A3RT"]:
    if i >= 30:
        a3rt.append(i)
a3rb = []
for i in df["A3RB"]:
    if i >= 30:
        a3rb.append(i)

#modified settings (arm position and loose tray)
a1lt2 = []
for i in df["A1LT2"]:
    if i >= 30:
        a1lt2.append(i)
a1lb2 = []
for i in df["A1LB2"]:
    if i >= 30:
        a1lb2.append(i)
a2lt2 = []
for i in df["A2LT2"]:
    if i >= 30:
        a2lt2.append(i)
a2lb2 = []
for i in df["A2LB2"]:
    if i >= 30:
        a2lb2.append(i)
a3lt2 = []
for i in df["A3LT2"]:
    if i >= 30:
        a3lt2.append(i)
a3lb2 = []
for i in df["A3LB2"]:
    if i >= 30:
        a3lb2.append(i)
a1rt2 = []
for i in df["A1RT2"]:
    if i >= 30:
        a1rt2.append(i)
a1rb2 = []
for i in df["A1RB2"]:
    if i >= 30:
        a1rb2.append(i)
a2rt2 = []
for i in df["A2RT2"]:
    if i >= 30:
        a2rt2.append(i)
a2rb2 = []
for i in df["A2RB2"]:
    if i >= 30:
        a2rb2.append(i)
a3rt2 = []
for i in df["A3RT2"]:
    if i >= 30:
        a3rt2.append(i)
a3rb2 = []
for i in df["A3RB2"]:
    if i >= 30:
        a3rb2.append(i)
```

```python
plt.figure(figsize=(20,10))
plt.plot(a1lt, color = "black", linewidth=1, label = "a1lt_production")
plt.plot(a1rt, color = "blue", linewidth=1, label = "a1rt_production")
plt.plot(a2lt, color = "red", linewidth=1, label = "a2lt_production")
plt.plot(a2rt, color = "grey", linewidth=1, label = "a2rt_production")
plt.plot(a3lt, color = "green", linewidth=1, label = "a3lt_production")
plt.plot(a3rt, color = "purple", linewidth=1, label = "a3rt_production")
axes = plt.gca()
axes.set_xlim([0,120])
axes.set_ylim([25,130])
plt.text(45, 137, "Data Analysed from:", color='black', size=12)
plt.text(60, 137, file_name, color='black', size=12)
plt.title("Temperature Profile - Top Thermocouple - Production Settings")
plt.suptitle("Balloon Length Failure Investigation")
plt.legend(loc = "upper left")
plt.xticks([])
plt.ylabel("Temperature [C]", color = "black")
plt.show()
```
